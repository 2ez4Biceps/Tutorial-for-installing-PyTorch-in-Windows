# PyTorch(GPU version) Installation Tutorial

### 1. Check the highest CUDA version supported by your graphics card.
11.6 here.Picture1.png






### 2. Find the corresponding version of PyTorch,CUDA and  cuDNN.
First check PyTorch official website https://pytorch.org/, current stable version is PyTorch 1.11.0 with CUDA 11.3.Picture2.png

### 3. Install CUDA 11.3.
Go to https://developer.nvidia.com/cuda-toolkit-archive and check archived versions of CUDA, find CUDA 11.3.0 and download it.

### 4. Install cuDNN.
Go to https://developer.nvidia.com/rdp/cudnn-archive and check archived versions of cuDNN. We have many choice and here install cuDNN v8.2.0 for CUDA 11.x.

### 5. Install PyTorch 1.11.0
Go to PyTorch official website https://pytorch.org/, get the install command. But actually this command will install the CPU version. So we need to go to PyTorch download library and find the version we need.My python version is 3.88, CUDA version is 11.3, so we download *cu113/torch-1.11.0%2Bcu113-cp38-cp38-win_amd64.whl* and *cu113/torchvision-0.12.0%2Bcu113-cp38-cp38-win_amd64.whl*. Here cu113 stands for CUDA version, cp38 stands for python version.
